# Tek Bir Nümerik Nitelik Üzerinde, Birden Çok Niteliği Gruplama ve Kümeleme

# agg() fonksiyonu format-1

    dataframe.groupby([gruplanacak_kategorik_nitelik1, gruplanacak_kategorik_nitelik2, ..]) \
    .kümelenecek_nümerik_nitelik \
    .agg(['kümeleme_fonk1','kümeleme_fonk2',...]) \
    .head()

# !!!!!!! Hem groupby() hem de agg() içine nitelikleri liste içinde vermeyi unutmayınız !!!!!!!

```python
# mesleklere göre ortalama gelir
meslek_grpby = df.groupby(["meslek"]).aylik_gelir \
.agg(['mean'])
```

```python
meslek_grpby.head(10)
```

```python
meslek_grpby.index
```

```python
meslek_grpby["meslek"] = meslek_grpby.index
```

```python
# Indeksi aynı zamanda bir sütun olarak df'e ekleyebiliriz.
meslek_grpby['meslek'] = meslek_grpby.index
meslek_grpby.head()
```

```python
# groupby içini liste halinde vermez isek çıktı Pandas.Series olur
```

```python
df_grpby_pd_series = df.groupby("meslek").aylik_gelir.agg('mean')
df_grpby_pd_series
```

# Pandas series dataframe yapma

```python
df_grpby_pd_series.reset_index().head()
```

# GroupBy Sonrası Indeksi Sıfırlama

```python
# mesleklere göre ortalama gelir
meslek_grpby2 = df.groupby(["meslek"]).aylik_gelir \
.agg(['mean']).reset_index()

meslek_grpby2.head()
```

# Meslekleri kazançlarına göre sıralama

```python
df_grpby_pd_series.reset_index().sort_values("meslek", ascending=False)
```

# Mesleklerin kazançlarını bar grafik ile çizmek

```python
import matplotlib.pyplot as plt
%matplotlib inline

df.groupby("meslek").aylik_gelir.mean() \
.sort_values(ascending=False) \
.plot(kind='bar')
plt.show()
```

# Birden fazla niteliği gruplama ve birden fazla nitelik üzerinde kümeleme fonksiyonu çalıştırma

```python
df.groupby(["meslek","sehir"]).aylik_gelir \
.agg(['count','mean','sum']) \
.sort_values("mean", ascending=False) \
.head()
```

# Birden fazla sıralama nasıl olur?

```python
df.sort_values(["yas","aylik_gelir"], ascending=[False,True]) \
.head(10)
```

```python
# 46 yaşındaki Berberlerin gelirleri de artan sıralı oldu.
```

# Birden Fazla Nümerik Değer Üzerinde groupby()

```python
df_meslek_sehir = df.groupby(["meslek","sehir"])[["yas","aylik_gelir"]] \
                    .agg(["mean","count"]) \
                    .reset_index()
df_meslek_sehir.head()
```
