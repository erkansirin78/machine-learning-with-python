<h2 id="drop_sutun">drop() ile sütun silmek</h2>

```python
df.drop('yas', axis=1, inplace=True)
```

```python
# Veri okuma esnasında da drop kullanabiliriz
df = pd.read_csv("D:/Datasets/simple_data.csv").drop('yas', axis=1)
df.head()
```

```python
df.head(20)
```

# Tekrarlayan satırları bir arada toplama duplicated()

```python
# tekrarlanan satırları görmek
duplicate_rows_df = df[ df.duplicated() ]
duplicate_rows_df.head()
```

# drop_duplicates()

```python
# Yukarıda 16. indekste bulunan satır tekrarlanıyor (15 ile tekrarlanmış)
df.drop_duplicates(inplace=True)
```

```python
duplicate_rows_df = df[df.duplicated()]
duplicate_rows_df.head()
```

# drop() ile satır silmek

```python
# 5 ve 7. satırları silmek
df.drop([5, 7], axis=0, inplace=True)
```
