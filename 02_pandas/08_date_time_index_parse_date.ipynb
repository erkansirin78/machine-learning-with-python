# tarih farkı bulma, string to datetime, promosyon tarihlerinde gerçekleşen satışlar

```python
# Promosyon tarih aralıkları
data = [['pro1', '2011-01-01', '2011-03-31'], ['pro2', '2011-04-01','2011-06-30'], ['pro3', '2011-07-01','2011-09-30'],
          ['pro4','2011-10-01','2011-12-31']
         ] 
  
# Promosyon dataframe
df_promotion = pd.DataFrame(data, columns = ['promotion', 'start_date', 'end_date']) 
   
df_promotion.head() 
```

```python
df_promotion.info()
```

```python
df_promotion['start_date'] = pd.to_datetime(df_promotion['start_date'])
df_promotion['end_date'] = pd.to_datetime(df_promotion['end_date'])
```

```python
df_promotion.info()
```

```python
df_promotion.head()
```

```python
def is_date_between_them(start_dates, end_dates, check_date):
    # bir tarihin iki tarih arasında olup olmadığını kontrol eden fonksiyon
    # Parametrelerde start_dates ve end_dates aynı dataframe e ait aralarında tarih aralığı bulunan tarihler
    flag = False 
    # tarih aralıklarının içinde dolaşalım ve belirtilen tarihin aralıkta olup olmadığını kontrol edelim
    for start_date, end_date in zip(start_dates, end_dates):
        # tarih aralıkta ise flag true olsun değilse false
        if( (check_date >= start_date) & (check_date <= end_date) ):
            flag = True
            break
    return flag
```

```python
# df e bir nitelik ekleyeceğimiz için onun uzunluğundaki bir listeyi paterne uygun dolduralım
is_promotion = []
# date sütunu içinde dolaş
for check_date in df.date:
    # eğer bu satırdaki tarih promotion tablosundaki tarihler arasında ise listeye 1 ekle yoksa 0
    if(is_date_between_them(df_promotion.start_date, df_promotion.end_date, check_date)):
        is_promotion.append(1)
    else: is_promotion.append(0)
```

```python
# df'e yeni sütun eklemeden önce boyut lar bunun için uygun mu kontrol edelim.
print(len(is_promotion))
print(df.shape)
```

```python
df["is_promotion"] = is_promotion
```
